```python
X_scaled = preprocessing.scale(X)
#y_scaled = y/9
y_scaled = y
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_scaled, y_scaled)
```

```python
X_train.shape
```

```python
y_train
```

```python
model = Sequential()
model.add(Dense(32, input_dim=X_train.shape[1], kernel_initializer='normal', activation='relu'))
#model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
#model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
#model.add(Dense(1))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adadelta', metrics=['mse','mae'])
```

```python
history = model.fit(X_train, y_train, epochs=150, batch_size=50,  verbose=1, validation_split=0.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
